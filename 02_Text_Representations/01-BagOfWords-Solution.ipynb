```python
import nltk
import numpy as np
import random
import string

from bs4 import BeautifulSoup
import urllib.request
import re
```

### Grab articles from wikipedia

```python
# Define URL and headers (pretend to be a real browser)
url = "https://en.wikipedia.org/wiki/Natural_language_processing"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}

# Create a request object with headers
req = urllib.request.Request(url, headers=headers)

# Open and read the page
with urllib.request.urlopen(req) as response:
    raw_html = response.read()

# Parse the HTML using BeautifulSoup
article_html = BeautifulSoup(raw_html, "lxml")

# Extract all paragraph tags
article_paragraphs = article_html.find_all("p")

# Combine all text into one string
article_text = ""
for para in article_paragraphs:
    article_text += para.get_text()

# Display first few hundred characters for verification
print(article_text[:500])
```

### Step 1: Tokenize

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Use a sentence tokenizer
corpus = nltk.sent_tokenize(article_text)
```

```python
corpus
```

```python
# remove punctuation and extra whitespaces and lowercase
for i in range(len(corpus )):
    corpus [i] = corpus [i].lower()
    corpus [i] = re.sub(r'\W',' ',corpus [i])
    corpus [i] = re.sub(r'\s+',' ',corpus [i])
```

```python
corpus
```

```python
print(corpus[20])
```

### Step 2: Create a dictionary of word frequency

```python
wordfreq = {}
for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
```

```python
wordfreq
```

```python
len(wordfreq)
```

### Remove the least frequent words using heapq library

```python
import heapq
most_freq = heapq.nlargest(200, wordfreq, key=wordfreq.get)
```

```python
most_freq
```

### Step 3: Create Bag of Words Model

```python
sentence_vectors = []
most_freq = list(most_freq)
most_freq_set = set(most_freq)

for sentence in corpus:
    # Tokenize and keep only relevant words (intersection)
    tokens = set(nltk.word_tokenize(sentence)) & most_freq_set

    # Binary vector: 1 if word present, else 0
    sent_vec = [1 if word in tokens else 0 for word in most_freq]

    sentence_vectors.append(sent_vec)
```

### Convert it to numpy array

```python
sentence_vectors = np.asarray(sentence_vectors)
```

## Implementation using Sklearn CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Step 1: Initialize CountVectorizer with top 200 most frequent words only
vectorizer = CountVectorizer(
    max_features=200,
    # stop_words='english',
    # token_pattern=r'\b[a-zA-Z]+\b'
)
```

```python
# Step 2: Fit and transform corpus
X = vectorizer.fit_transform(corpus)
```

```python
# Step 3: Get feature names (vocabulary)
vocab = vectorizer.get_feature_names_out()
vocab
```

```python
# Step 4: Convert BoW matrix to a pandas DataFrame
bow_df = pd.DataFrame(X.toarray(), columns=vocab)
bow_df.head()
```

```python
print("Words in vocab but not in most_freq:", set(vocab) - set(most_freq))
print("Words in most_freq but not in vocab:", set(most_freq) - set(vocab))
```
