### Import the libraries

```python
import spacy
import textacy
from textacy.extract import keyterms
```

### Load a spacy model, which will be used for all further processing

```python
en = spacy.load("en_core_web_sm")
```

### Load a sample text to find keywords
- kpe_sample_text.txt

```python
mytext = open('kpe_sample_text.txt').read()
mytext
```

### Convert the text into a spacy document

```python
doc = textacy.make_spacy_doc(mytext, lang=en) 
doc
```

## Find keywords

## 1. Noun Chunks

```python
print([chunk for chunk in textacy.extract.noun_chunks(doc)])
```

### Issues ???
- Noun chunks don't filter for importance. 

## 2. TextRank

```python
keyterms.textrank(doc, topn=5)
```

### Get more keywords

```python
keyterms.textrank(doc, topn=20)
```

### Keywords using TextRank algorithm

```python
[kps for kps, weights in keyterms.textrank(doc, normalize="lemma", topn=10)]
```

### Keywords using SGRank algorithm

```python
[kps for kps, weights in keyterms.sgrank(doc, topn=10)]
```

Issue: **Overlapping key phrases**

Solution: **aggregage_term_variants**
- Choosing one of the grouped terms per item will give us a list of non-overlapping key phrases

```python
from textacy.extract import utils
# Extract key terms with SGRank
terms = {term for term, weight in keyterms.sgrank(doc)}

# Aggregate variants (e.g., plural/singular, different cases)
aggregated = utils.aggregate_term_variants(terms)
aggregated
```
