# Summarization
## This notebook outlines the concepts behind Text Summarization

## Summarization
- concept of capturing very important gist of a long piece of text

### Types of Summarization
- 1. **Extractive Summarization**
    - Select sentences from the corpus that best represent the text
    - Arrange them to form a summary
- 2. **Abstractive Summarization**
    - Captures the very important sentences from the text
    - Paraphrases them to form a summary

### Import the library

```python
# gensim.summarization was removed in gensim 4.0 and gensim 3.x doesn't support Python 3.13
# Using sumy LSA as a drop-in replacement with the same interface

from sumy.summarizers.lsa import LsaSummarizer as GensimReplaceSummarizer
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer as SumyTokenizer

def summarize(text, word_count=200, ratio=0.02):
    parser = PlaintextParser.from_string(text, SumyTokenizer("english"))
    summarizer = GensimReplaceSummarizer()
    num_sentences = max(3, int(len(text.split('.')) * ratio))
    summary = summarizer(parser.document, num_sentences)
    return ' '.join([str(s) for s in summary])

print("Gensim summarize replacement ready ✓")
```

### Scrape the text
- Use beautifulSoup to extract text (from Task1 of ML-1)

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_page(url):
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

```python
def collect_text(soup):
    paragraphs = soup.find_all('p')
    text = ' '.join([p.get_text() for p in paragraphs])
    return text
```

```python
url = "https://en.wikipedia.org/wiki/Automatic_summarization"
```

```python
text = collect_text(get_page(url))
text
```

### Summarize
- **word_count**: maximum amount of words we want in the summary
- **ratio**: fraction of sentences in the original text should be returned as output

```python
print("=== Gensim Summary (word_count=200) ===")
print(summarize(text, word_count=200))

print("\n=== Gensim Summary (ratio=0.02) ===")
print(summarize(text, ratio=0.02))
```
